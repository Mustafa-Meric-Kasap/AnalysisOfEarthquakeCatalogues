```python
import pandas as pd

df_without_aftershocks = pd.read_csv("../../data/cleaned/usgs_pacific_belt_93_now_rectangular_region_aftershocks_filtered.csv")
df_without_aftershocks
```

```python
df_without_aftershocks['Datetime'] = pd.to_datetime(df_without_aftershocks['Datetime'], format="%d/%m/%Y %H:%M:%S")  # Ensure it's in datetime format
df_without_aftershocks
```

```python
df_new = df_without_aftershocks.copy()
df_new = df_new[
    ((-60 <= df_new['Latitude']) & (df_new['Latitude'] <= 60) & (-130 <= df_new['Longitude']) & (df_new['Longitude'] <= -70)) |
    ((-60 <= df_new['Latitude']) & (df_new['Latitude'] <= 60) & (120 <= df_new['Longitude']) & (df_new['Longitude'] <= 180))]
df_new
```

```python
df_new['Datetime'] = pd.to_datetime(df_new['Datetime'], format="%d/%m/%Y %H:%M:%S")  # Ensure it's in datetime format
df_new
```

```python
from src.data import preprocess
import torch
df_new2 = df_new.copy()
df_new2 = df_new2[df_new2['Datetime'] > '2000-01-01']
X_aftershock_filtered_label0, y_aftershock_filtered_label0 = preprocess.create_classification_data(df_without_aftershocks, df_new2)
torch.sum(y_aftershock_filtered_label0 == 0).item()
```

```python
import numpy as np
X_aftershock_filtered_label1 = torch.load("../../data/raw/X_tensor_pacific_belt_aftershock_filtered_label1.pt")
y_aftershock_filtered_label1 = torch.load("../../data/raw/y_tensor_pacific_belt_aftershock_filtered_label1.pt")

X_combined = np.concatenate((X_aftershock_filtered_label1, X_aftershock_filtered_label0), axis=0)
y_combined = np.concatenate((y_aftershock_filtered_label1, y_aftershock_filtered_label0), axis=0)
shuffle_indices = np.random.permutation(X_combined.shape[0])

X_shuffled = X_combined[shuffle_indices]
y_shuffled = y_combined[shuffle_indices]

X_tensor = torch.from_numpy(X_shuffled)
y_tensor = torch.from_numpy(y_shuffled)
X_tensor.size()
```
