```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load tensors from files
time_difference_tensor = torch.load('/content/time_difference_tensor.pt')
time_difference_tensor_aftershocks_filtered = torch.load('/content/time_difference_tensor_aftershocks_filtered.pt')

# Convert tensors to numpy arrays
time_difference_array = time_difference_tensor.numpy()
time_difference_filtered_array = time_difference_tensor_aftershocks_filtered.numpy()

# Flatten the arrays for frequency calculation
time_difference_flat = time_difference_array.flatten()
time_difference_filtered_flat = time_difference_filtered_array.flatten()

# Create frequency tables
freq_table_time_difference = pd.value_counts(time_difference_flat).sort_index()
freq_table_filtered = pd.value_counts(time_difference_filtered_flat).sort_index()

# Convert to DataFrames for better visualization
df_time_difference = freq_table_time_difference.reset_index()
df_time_difference.columns = ['Value', 'Frequency']

df_filtered = freq_table_filtered.reset_index()
df_filtered.columns = ['Value', 'Frequency']

# Print frequency tables
print("Frequency Table for Time Difference Tensor:")
print(df_time_difference)

print("\nFrequency Table for Time Difference Tensor Aftershocks Filtered:")
print(df_filtered)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(time_difference_flat, bins=50, color='blue', edgecolor='black')
plt.title('Histogram of Time Difference Tensor')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot histogram for Time Difference Tensor Aftershocks Filtered
plt.subplot(1, 2, 2)
plt.hist(time_difference_filtered_flat, bins=50, color='blue', edgecolor='black')
plt.title('Histogram of Time Difference Tensor Aftershocks Filtered')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Compute statistics for Time Difference Tensor
mean_time_difference = np.mean(time_difference_flat)
std_dev_time_difference = np.std(time_difference_flat)
variance_time_difference = np.var(time_difference_flat)
min_time_difference = np.min(time_difference_flat)
max_time_difference = np.max(time_difference_flat)

# Compute statistics for Time Difference Tensor Aftershocks Filtered
mean_filtered = np.mean(time_difference_filtered_flat)
std_dev_filtered = np.std(time_difference_filtered_flat)
variance_filtered = np.var(time_difference_filtered_flat)
min_filtered = np.min(time_difference_filtered_flat)
max_filtered = np.max(time_difference_filtered_flat)

# Print the results
print("Statistics for Time Difference Tensor:")
print(f"Mean: {mean_time_difference:.2f} hours")
print(f"Standard Deviation: {std_dev_time_difference:.2f} hours")
print(f"Variance: {variance_time_difference:.2f} hours^2")
print(f"Minimum: {min_time_difference:.2f} hours")
print(f"Maximum: {max_time_difference:.2f} hours")

print("\nStatistics for Time Difference Tensor Aftershocks Filtered:")
print(f"Mean: {mean_filtered:.2f} hours")
print(f"Standard Deviation: {std_dev_filtered:.2f} hours")
print(f"Variance: {variance_filtered:.2f} hours^2")
print(f"Minimum: {min_filtered:.2f} hours")
print(f"Maximum: {max_filtered:.2f} hours")
```
