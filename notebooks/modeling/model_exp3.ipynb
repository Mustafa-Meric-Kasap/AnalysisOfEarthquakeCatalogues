```python
def lat_lon_to_cartesian(lat, lon, R=6371.0):
    # Convert latitude and longitude from degrees to radians
    lat_rad = np.radians(lat)
    lon_rad = np.radians(lon)
    
    # Calculate x, y, z coordinates
    x = R * np.cos(lat_rad) * np.cos(lon_rad)
    y = R * np.cos(lat_rad) * np.sin(lon_rad)
    z = R * np.sin(lat_rad)
    
    return np.array([x, y, z])

def spherical_vector(lat1, lon1, lat2, lon2):
    # Convert points to Cartesian coordinates
    p1 = lat_lon_to_cartesian(lat1, lon1)
    p2 = lat_lon_to_cartesian(lat2, lon2)
    
    # Calculate Cartesian difference vector
    d_cartesian = p2 - p1
    
    # The resulting vector contains x, y, and z components
    return d_cartesian
```

```python
def lat_lon_to_cartesian_vectorized(df, R=6371.0):
    lat_rad = np.radians(df['Latitude'].values)
    lon_rad = np.radians(df['Longitude'].values)
    
    x = R * np.cos(lat_rad) * np.cos(lon_rad)
    y = R * np.cos(lat_rad) * np.sin(lon_rad)
    z = R * np.sin(lat_rad)
    
    df_copy = df.copy()
    df_copy['X'] = x
    df_copy['Y'] = y
    df_copy['Z'] = z
    
    return df_copy
```

```python
from src.util.math_utils import date_time_diff_in_hours

def get_cartesian_properties(df, idx_array):
    event_id = df.iloc[idx_array]["Event ID"]
    magnitude = df.iloc[idx_array]["xM"]
    depth = df.iloc[idx_array]["Depth(km)"]
    x = df.iloc[idx_array]["X"]
    y = df.iloc[idx_array]["Y"]
    z = df.iloc[idx_array]["Z"]
    date = df.iloc[idx_array]["Datetime"]
    return np.array([event_id, magnitude, depth, x, y, z, date])


def get_properties_all(df):
    arr = []
    for i in range(len(df) - 1):
        curr_elt = get_cartesian_properties(df, i)
        next_elt = get_cartesian_properties(df, i + 1)
        mag_diff = next_elt[1] - curr_elt[1]
        depth_diff = next_elt[2] - curr_elt[2]
        x_vec = next_elt[3] - curr_elt[3]
        y_vec = next_elt[4] - curr_elt[4]
        z_vec = next_elt[5] - curr_elt[5]
        time_diff = date_time_diff_in_hours(next_elt[6], curr_elt[6])
        arr.append(np.array([mag_diff, depth_diff, x_vec, y_vec, z_vec, time_diff]))
    return np.array(arr)
```

```python
from src.util.earthquake_filter_utils import magnitude_filter
from src.util.earthquake_filter_utils import distance_filter
from src.util.earthquake_filter_utils import past_earthquakes_filter
from src.util.earthquake_filter_utils import time_filter
from src.util.earthquake_filter_utils import future_earthquakes_filter

def comprehensive_filter(df, event_id, radius_km=10, years=30, num_earthquakes=30):
    # Filter by magnitude first
    mag_filtered_df = magnitude_filter(df, 3.0)

    # Apply distance filter
    distance_filtered_df = distance_filter(mag_filtered_df, event_id, radius_km)

    # Apply time filter
    time_filtered_df = time_filter(distance_filtered_df, event_id, years)

    # Get past and future earthquakes
    past_eqs = past_earthquakes_filter(time_filtered_df, event_id, num_earthquakes)
    future_eqs = future_earthquakes_filter(time_filtered_df, event_id, num_earthquakes)
    
    # Combine past and future earthquakes
    result = pd.concat([past_eqs.iloc[:-1], future_eqs[0:]], ignore_index=True)

    return result
```

```python
def calculate_aftershock_duration_hours(large_eq_mag):
    return 10 ** (large_eq_mag - 3.0) * 24 
```

```python
def filter_large_eqs_before_after(df, radius_km=10, years=30, num_earthquakes=30, magnitude=5.5):
    res_df = pd.DataFrame(columns=df.columns)
    large_earthquakes = earthquake_filter_utils.magnitude_filter(df, magnitude)
    # Traverse over all large earthquakes
    for i in range(len(large_earthquakes)):
        large_eq_eid = large_earthquakes.iloc[i]['Event ID'] # Get large_earthquake i event id
        # create a df including large earthquake and num_earthquakes earthquakes before and after it. 
        filtered = comprehensive_filter(df, large_eq_eid, radius_km, years, num_earthquakes)  
        res_df = pd.concat([res_df, filtered], ignore_index=True)
        
    # drop duplicate earthquakes with looking their unique event id
    res_df = res_df.drop_duplicates(subset=['Event ID'], inplace=False)
    return res_df
```

```python
def remove_elements(df1, df2, key_column='Event ID'):
    """
    Removes all rows from df1 that are present in df2 based on the key_column.
    
    Parameters:
        df1 (pd.DataFrame): The main DataFrame from which rows are to be removed.
        df2 (pd.DataFrame): The DataFrame containing rows to remove from df1.
        key_column (str): The column to compare between df1 and df2.
        
    Returns:
        pd.DataFrame: A new DataFrame with the rows in df2 removed from df1.
    """
    # Use ~isin to filter out rows in df1 where the key_column exists in df2
    df_filtered = df1[~df1[key_column].isin(df2[key_column])]
    
    return df_filtered
```
